## Changing the data column names for better understanding

```python
data1.rename(columns={'Region':'State'},inplace=True)
data2.rename(columns={'Region':'State','Region.1':'Region','longitude':'Longitude','latitude':'Latitude'},inplace=True)
print(data1.columns)
print(data2.columns)
```

## Correcting the datatypes and adding a new column

```python
data1.info()
```

```python
import datetime as dt
import calendar as cal
```

```python
data1[' Date']=pd.to_datetime(data1[' Date'],dayfirst=True)
data1['month_int']=data1[' Date'].dt.month
data1['month']=data1['month_int'].apply(lambda x:cal.month_abbr[x])
```

```python
data2[' Date']=pd.to_datetime(data2[' Date'],dayfirst=True)
data2['month_int']=data2[' Date'].dt.month
data2['month']=data2['month_int'].apply(lambda x:cal.month_abbr[x])
```

## Dropping the columns which are not required

```python
data1.columns
```

```python
data2.drop([' Frequency','Longitude', 'Latitude','month_int'],inplace=True,axis=1)
data2.head()
```

```python
data1.drop([' Frequency','month_int'],inplace=True,axis=1)
data1.head()
```

## Data Visualizations

**Top 10 states with the highest unemployment rate in India and till 11_2020**

**Top 10 states with the highest employment rate in India and till 11_2020**

*  The employment rate sum of Telegana decreased and that of Andhra Pradesh increased after 11_2020.

*  Average Labor participation in Karnataka and Maharashtra decreased and Himachal Pradesh and Chattisgarh increased after 11_2020.
