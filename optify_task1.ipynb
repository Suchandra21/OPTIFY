# **IRIS DATASET**

```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
data=pd.read_csv("/content/iris_flower.csv")
data
```

```python
#finding the number of columns and rows
data.shape
```

```python
#to find the columns
print(data.columns)
```

```python
#finding the statistical information (mean,median,mode) from the dataset
data.describe()
```

```python
#finding the total number of sepal and petal length
data.info()
```

So, we conclude that the dataset has no null values and the datatypes are also correct.

```python
#to check whether any '0' value exists or not
dt=data.drop('species',axis=1)
sns.heatmap(dt)
```

We conclude that there are few values of petal_width near about 0 or maybe are 0.So, to confirm we do the next code.

```python
dt[(dt['petal_width']<1) & (dt['petal_width']>0)].any()
```

We conclude that the values are less than 1 but more than 0. So they are near about zero.

```python
# to see the number of species
plt.bar(data['species'].value_counts().index, data.groupby('species')['sepal_length'].mean(),color=['red','green','pink'])
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.title('Mean Sepal Length by Species')
plt.show()
```

So we can see that Virginica has higher sepal length than the other two species.  

```python
plt.bar(data['species'].value_counts().index,data.groupby('species')['sepal_width'].mean(),color=['red','green','pink'])
plt.xlabel("Species")
plt.ylabel("Mean Sepal width")
plt.title("Mean sepal width by species")
```

```python
plt.bar(data['species'].value_counts().index,data.groupby('species')['petal_length'].mean(),color=['red','green','pink'])
plt.xlabel("Species")
plt.ylabel("Mean petal length")
plt.title("Mean petal length by species")
```

```python
plt.bar(data['species'].value_counts().index,data.groupby('species')['petal_width'].mean(),color=['red','green','pink'])
plt.xlabel("Species")
plt.ylabel("Mean petal width")
plt.title("Mean petal width by species")
```

```python
sset=data[data['species']=='Iris-setosa'].value_counts().sum()
vir=data[data['species']=='Iris-virginica'].value_counts().sum()
ver=data[data['species']=='Iris-versicolor'].value_counts().sum()
print("Setosa count:",sset)
print("Virginica count:",vir)
print("Versicolor count:",ver)
```

So we have got equal number of flowers from each species.

```python
# relationship between sepal_length and petal_length w.r.t. species
sns.scatterplot(x=data['petal_length'],y=data['sepal_length'],hue='species',data=data)
plt.title("Relation between sepal_length and petal_length")
plt.show()
```

Here we see unlike the versicolor and virgnica , the sepal length of flowers of sertosa species is not depended on the petal length.  

```python
# to see how different factors depend upon each other
sns.pairplot(data,hue='species')
```

In case of species setosa, its sepal length,petal length and petal width mostly depend on sepal width. In the rest virginica and versicolor depend more.
